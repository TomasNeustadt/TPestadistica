### **1. Importación de librerías**

En esta sección se cargan las librerías necesarias para realizar el análisis computacional:
- `yfinance`: descarga de datos.
- `numpy` y `pandas`: cálculos numéricos.
- `matplotlib` y `seaborn`: representacion gráfica.

```python
import yfinance as yf
import datetime
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

pd.options.display.float_format = "{:,.4f}".format
plt.style.use("seaborn-v0_8")
```

### **2. Definición de funciones auxiliares**

```python
def descargar_precios(tickers, años=4):
    end = datetime.datetime.now()
    start = end - datetime.timedelta(days=365*años)
    data = yf.download(tickers, start=start, end=end)
    return data["Close"].rename(columns={"^GSPC": "SP500", "^VIX": "VIX"})


def calcular_estadisticos(rend):
    media = rend.mean() * 100
    var = rend.var() * (100**2)
    sd = rend.std() * 100

    return pd.DataFrame({
        "Promedio Muestral (%)": media,
        "Varianza Muestral (%^2)": var,
        "Desvío Estándar (Volatilidad) (%)": sd
    })


def scatter(x, y, label_x, label_y, title):
    plt.figure(figsize=(9, 6))
    sns.scatterplot(x=x * 100, y=y * 100, alpha=0.5)
    plt.axhline(0, color='red', linestyle='--', linewidth=1)
    plt.axvline(0, color='red', linestyle='--', linewidth=1)
    plt.title(title, fontsize=15)
    plt.xlabel(label_x + " (%)")
    plt.ylabel(label_y + " (%)")
    plt.grid(linestyle='--', alpha=0.5)
    plt.show()


def construir_portafolios(media1, media2, var1, var2, cov12, nombre1, nombre2):
    w1 = np.linspace(0, 1, 51)
    w2 = 1 - w1

    retorno = w1 * media1 + w2 * media2

    var_p = (w1**2 * var1) + (w2**2 * var2) + (2 * w1 * w2 * cov12)
    vol = np.sqrt(var_p) * 100

    df = pd.DataFrame({
        f"Peso {nombre1} (%)": w1 * 100,
        f"Peso {nombre2} (%)": w2 * 100,
        "Rendimiento Esperado (%)": retorno,
        "Volatilidad (Riesgo) (%)": vol
    })

    return df


def frontera(df, nombre1, nombre2):
    mvp = df.loc[df["Volatilidad (Riesgo) (%)"].idxmin()]

    plt.figure(figsize=(11, 7))
    sns.scatterplot(
        x="Volatilidad (Riesgo) (%)",
        y="Rendimiento Esperado (%)",
        data=df,
        s=60
    )

    plt.scatter(
        mvp["Volatilidad (Riesgo) (%)"],
        mvp["Rendimiento Esperado (%)"],
        s=200,
        color="red",
        edgecolor="black",
        label="Mínima Varianza"
    )

    plt.title(f"Frontera Eficiente: {nombre1} vs {nombre2}", fontsize=16)
    plt.xlabel("Volatilidad (%)")
    plt.ylabel("Retorno Esperado (%)")
    plt.grid(linestyle='--', alpha=0.7)
    plt.legend()
    plt.show()

    return mvp
```

### **3. Descarga de precios y cálculo de rendimientos diarios**

Se descargan los precios de cierre diarios de los últimos 4 años para:
- SP500
- VIX
- BABA
- EBAY

Luego, se calculan los rendimientos logarítmicos ya que es el tipo de rendimiento estandar utilizado para los analisis financieros.


```python
tickers = ["^GSPC", "^VIX", "BABA", "EBAY"]

precios = descargar_precios(tickers)
rend = np.log(precios).diff().dropna()

rend.head()
```

### **4. Estadísticas descriptivas**

Se calcula para cada activo:

- Promedio muestral del rendimiento diario.
- Varianza muestral.
- Desvío estándar (volatilidad).

Estas medidas permiten comparar riesgo y retorno de los activos.


```python
estad = calcular_estadisticos(rend)
estad
```

### **5. Ranking de activos según retorno y volatilidad**

Ordenamos los activos de mayor a menor:
- En rendimiento promedio.
- En volatilidad.

Esto permite identificar activos más riesgosos y más rentables.


```python
estad.sort_values("Promedio Muestral (%)", ascending=False)
estad.sort_values("Desvío Estándar (Volatilidad) (%)", ascending=False)
```

### **6. Boxplot de rendimientos ordenado por rango intercuartil (IQR)**

El IQR mide la dispersión central de los datos, menos sensible a outliers que la varianza.


```python
IQR = (rend.quantile(0.75) - rend.quantile(0.25)).sort_values(ascending=False)

rend_melt = rend.melt(var_name="Activo", value_name="Rendimiento Diario (%)")
rend_melt["Rendimiento Diario (%)"] *= 100

plt.figure(figsize=(10, 6))
sns.boxplot(
    x="Activo", 
    y="Rendimiento Diario (%)",
    data=rend_melt,
    order=IQR.index
)
plt.title("Boxplot de Rendimientos Ordenado por IQR", fontsize=15)
plt.grid(axis='y', linestyle='--', alpha=0.4)
plt.show()
```

### Relación entre el rango intercuartil y otras estadísticas descriptivas

El rango intercuartil (IQR) está directamente relacionado con la variabilidad de los datos,
igual que la varianza y el desvío estándar. Mientras que la varianza y la desviación estándar
incorporan toda la dispersión de la muestra (incluyendo outliers), el IQR mide exclusivamente
la dispersión central del 50% de los datos. Por eso suele coincidir en ordenamiento con los
activos que tienen mayor volatilidad, aunque es más robusto frente a valores extremos.


### **7. Análisis de correlación: SP500–VIX y BABA–EBAY**

Se generan diagramas de dispersión para visualizar la relación entre pares de activos.

Luego se calcula la covarianza muestral para evaluar el signo de la dependencia.


```python
mat_cov = rend.cov()

# SP500 vs VIX
scatter(rend["SP500"], rend["VIX"],
        "SP500", "VIX", "SP500 vs VIX – Rendimientos")
print("Covarianza SP500–VIX =", mat_cov.loc["SP500", "VIX"])

# BABA vs EBAY
scatter(rend["BABA"], rend["EBAY"],
        "BABA", "EBAY", "BABA vs EBAY – Rendimientos")
print("Covarianza BABA–EBAY =", mat_cov.loc["BABA", "EBAY"])
```

### **8. Construcción de portafolios entre BABA y EBAY**

Se generan 51 portafolios variando el peso de BABA entre 0 y 1.

Se calcula:
- Rendimiento esperado del portafolio.
- Volatilidad del portafolio.


```python
df_baba_ebay = construir_portafolios(
    estad.loc["BABA","Promedio Muestral (%)"],
    estad.loc["EBAY","Promedio Muestral (%)"],
    rend["BABA"].var(), rend["EBAY"].var(),
    mat_cov.loc["BABA","EBAY"],
    "BABA", "EBAY"
)

df_baba_ebay.head()
```

### ¿Era necesario calcular los 51 portafolios para elegir el de mayor rendimiento?

No. El rendimiento esperado de un portafolio es una combinación convexa de las medias:

Esto implica que el portafolio con mayor rendimiento esperado es siempre aquel que invierte el
100% en el activo con mayor media. Por lo tanto, la recomendación de mayor rentabilidad se puede
obtener simplemente comparando los promedios muestrales, sin necesidad de generar toda la grilla
de portafolios.


### **9. Frontera eficiente entre BABA y EBAY**

Se grafica la frontera eficiente de riesgo–retorno
y se identifica el portafolio de mínima varianza (MVP).


```python
mvp_baba_ebay = frontera(df_baba_ebay, "BABA", "EBAY")
mvp_baba_ebay
```

### Independencia entre activos

En los pares analizados:

- La covarianza SP500–VIX es negativa.
- La covarianza BABA–EBAY es positiva.

En ambos casos la covarianza es distinta de cero, lo cual demuestra que las variables NO son
independientes. Además, los diagramas de dispersión muestran patrones definidos: relación
negativa marcada entre SP500–VIX y relación positiva moderada entre BABA–EBAY. Esto confirma
que existe dependencia estadística entre los activos.


### **10. Portafolios entre SP500 y VIX**

Repetimos todo el análisis de portafolios,
evaluando cómo cambia la frontera eficiente cuando se usan un índice de mercado y el índice de volatilidad.


```python
df_sp_vix = construir_portafolios(
    estad.loc["SP500","Promedio Muestral (%)"],
    estad.loc["VIX","Promedio Muestral (%)"],
    rend["SP500"].var(), rend["VIX"].var(),
    mat_cov.loc["SP500","VIX"],
    "SP500", "VIX"
)

mvp_sp_vix = frontera(df_sp_vix, "SP500", "VIX")
mvp_sp_vix
```
